```python
def Max_Of_Two(a: int, b: int) -> int:
    return a if a > b else b
```

```python
from typing import Tuple

def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a
```

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def is_perfect_number(n: int) -> bool:
    if n <= 0:
        return False
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
```

```python
def find_index(lst: list, k: int) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
```

```python
def factorial(n: int) -> int:
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res
```

```python
def calculator(num1: float, num2: float, operat: str) -> float:
    if operat == '+':
        res = num1 + num2
    elif operat == '-':
        res = num1 - num2
    elif operat == '*':
        res = num1 * num2
    elif operat == '/':
        res = num1 / num2 if num2 != 0 else 0 
    return round(res, 2)
```

```python
def sum_of_digits(n: int) -> int:
    tong = 0
    n = abs(n)
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
```

```python
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    
    mapping_a_to_b = {}
    mapping_b_to_a = {}
    
    for char_a, char_b in zip(a, b):
        if char_a in mapping_a_to_b:
            if mapping_a_to_b[char_a] != char_b:
                return False
        else:
            mapping_a_to_b[char_a] = char_b
        if char_b in mapping_b_to_a:
            if mapping_b_to_a[char_b] != char_a:
                return False
        else:
            mapping_b_to_a[char_b] = char_a
            
    return True
```

```python
def hamming_distance(x: int, y: int) -> int:
    diff = x ^ y
 
    count = 0
    while diff > 0:
        count += diff & 1
        diff >>= 1
    return count
```
