```python
a = int(input("Enter a positive integer a: "))

print(f"Multiplication table for {a}:")
for i in range(1, 11):
    print(f"{a} x {i} = {a * i}")
```

```python
n = int(input("Enter a positive integer n: "))
S = 0
for i in range(1, n + 1):
    S += i

print(f"The sum S is: {S}")
```

```python
line = input("Enter n, m, k: ").split()
n, m, k = int(line), int(line[1]), int(line[4])

matrix = []
print(f"Enter {n} rows of {m} integers:")
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

if k < n:
    row_sum = sum(matrix[k])
    print(f"Sum of elements in row {k}: {row_sum}")

if k < m:
    col_sum = sum(matrix[i][k] for i in range(n))
    print(f"Sum of elements in column {k}: {col_sum}")
```

```python
a = int(input("Nhập a: "))
count = 0
for i in range(1, a + 1):
    if a % i == 0:
        count += 1
print("Số ước:", count)
```

```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

```python
a = int(input("Nhập a: "))
check = False
for i in range(a + 1):
    if i * i == a:
        check = True
        break
print("Là số chính phương" if check else "Không phải số chính phương")
```

```python
max_val = -float('inf')
min_val = float('inf')
while True:
    n = int(input("Nhập số (-1 để dừng): "))
    if n == -1: break
    if n > max_val: max_val = n
    if n < min_val: min_val = n
print(f"Max: {max_val}, Min: {min_val}")
```
