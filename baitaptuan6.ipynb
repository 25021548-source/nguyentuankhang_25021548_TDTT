```python
lst = list(map(int, input().split()))
seen = set()
result = []
for x in lst:
    if x not in seen:
        result.append(x)
        seen.add(x)
print(result)
```

```python
lst = list(map(int, input().split()))
cumulative_sum = []
current_sum = 0
for x in lst:
    current_sum += x
    cumulative_sum.append(current_sum)
print(cumulative_sum)
```

```python
tup = tuple(map(int, input().split()))
k = int(input())
if len(tup) == 0:
    print(())
else:
    k = k % len(tup)
    res = tup[k:] + tup[:k]
    print(res)
```

```python
data = input().split()
d = {}
for item in data:
    key, value = item.split(':')
    if key not in d:
        d[key] = []
    d[key].append(value)
print(d)
```

```python
lst = list(map(int, input().split()))
res = {'positives': 0, 'negatives': 0, 'zeros': 0}
for x in lst:
    if x > 0: res['positives'] += 1
    elif x < 0: res['negatives'] += 1
    else: res['zeros'] += 1
print(res)
```

```python

strings = input().split()
freq = {}
for s in strings:
    freq[s] = freq.get(s, 0) + 1
print(freq)
```

```python
def get_dict():
    d = {}
    for item in input().split():
        k, v = item.split(':')
        d[k] = int(v)
    return d

dict1 = get_dict()
dict2 = get_dict()

for k, v in dict2.items():
    dict1[k] = dict1.get(k, 0) + v
```

```python
lst = list(map(int, input().split()))
k = int(input())
pairs = []
for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] + lst[j] == k:
            pairs.append((lst[i], lst[j]))
```

```python
lst = list(map(int, input().split()))
if not lst: exit()
freq = {}
for x in lst:
    freq[x] = freq.get(x, 0) + 1

max_f = max(freq.values())
candidates = [k for k, v in freq.items() if v == max_f]
print(min(candidates))
```
