```python
import heapq

def find_kth_largest(matrix, k):
    flat_list = [item for sublist in matrix for item in sublist]
    return heapq.nlargest(k, flat_list)[-1]
```

```python
def solve_w9a4(a, b):
    i, j = 0, 0
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            i += 1
            j += 1
        elif a[i].isupper():
            return "NO"
        else:
            i += 1
    while i < len(a):
        if a[i].isupper():
            return "NO"
        i += 1
    return "YES" if j == len(b) else "NO"
```

```python
def is_sub_sequence(list1, list2):
    n, m = len(list1), len(list2)
    for i in range(m - n + 1):
        if list2[i : i + n] == list1:
            return "YES"
    return "NO"
```

```python
def next_permutation(n, a):
    i = n - 2
    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1
    if i == -1:
        return sorted(a)
    
    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
    return a
```

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'U': y += 1
        elif move == 'D': y -= 1
        elif move == 'L': x -= 1
        elif move == 'R': x += 1
        
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
```

```python
def Median(nums):
    a, b, c, d, e = nums
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b 
    if e < c:
        if e < a: e, a, c = c, a, e
        else: e, c = c, e
    else:
        if e > d: e, d = d, e
    if b < c:
        return b if b > e else (e if e < c else c)
    else:
        return c if c > e else (e if e < b else b)
```

```python
import bisect

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    r = 0
    for h in houses:
        idx = bisect.bisect_left(heaters, h)
        dist_right = heaters[idx] - h if idx < len(heaters) else float('inf')
        dist_left = h - heaters[idx-1] if idx > 0 else float('inf')
        r = max(r, min(dist_left, dist_right))
    return r
```
