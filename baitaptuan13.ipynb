```python
def is_perfect(n):
    if n <= 0:
        return False
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n

n = int(input())
print(is_perfect(n))
```

```python
a = list(map(int, input().split()))
if not a:
    print(0)
elif len(a) == 1:
    print(a[0])
else:
    dp = [0]*len(a)
    dp[0] = a[0]
    dp[1] = max(a[0], a[1])
    for i in range(2, len(a)):
        dp[i] = max(dp[i-1], dp[i-2] + a[i])
    print(dp[-1])
```

```python
def count(path, character):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            return f.read().count(character)
    except FileNotFoundError:
        return -1
```

```python
n = int(input())
best_name = ""
best_score = -1

for _ in range(n):
    name, score = input().rsplit(" ", 1)
    score = float(score)
    if score > best_score:
        best_score = score
        best_name = name

print(best_name)
```
