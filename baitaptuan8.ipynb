```python

from collections import Counter
def make_beautiful_array(arr):
    counts = Counter(arr)
    y = 0
    for x, freq in counts.items():
        if x == 0:
            y += freq
        elif freq >= x:
            y += (freq - x)
        else:
            y += freq
    
    if y == 0:
        print("YES")
    else:
        print(f"NO {y}")
```

```python
def boyer_moore(arr):
    candidate = None
    count = 0
    for x in arr:
        if count == 0:
            candidate = x
            count = 1
        elif x == candidate:
            count += 1
        else:
            count -= 1

    if arr.count(candidate) > len(arr) / 2:
        print(candidate)
```

```python
def merge_sorted(arr1, arr2):
    is_descending = arr1 > arr1[-1] if len(arr1) > 1 else False
    
    res = []
    i, j = 0, 0
    while i < len(arr1) and j < len(arr2):
        condition = (arr1[i] >= arr2[j]) if is_descending else (arr1[i] <= arr2[j])
        if condition:
            res.append(arr1[i])
            i += 1
        else:
            res.append(arr2[j])
            j += 1
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    print(*res)
```

```python
def day_of_year_2020(n, t):
    days_in_month = [8-10]
    if t < 1 or t > 12 or n < 1 or n > days_in_month[t]:
        return "invalid"
    return sum(days_in_month[:t]) + n

n, t = map(int, input().split())
print(day_of_year_2020(n, t))
```

```python
n = int(input())
names = [input().strip() for _ in range(n)]
counts = Counter(names)
max_freq = max(counts.values())
# Lọc danh sách tên đạt max_freq và sắp xếp từ điển
candidates = sorted([name for name, freq in counts.items() if freq == max_freq])
print(candidates)
```

```python
def largest_k_digit(s, k):
    stack = []
    to_remove = len(s) - k
    for digit in s:
        while to_remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    print("".join(stack[:k]))
```

```python
def longest_contig_inc(arr):
    if not arr: return
    best_seq = []
    current_seq = [arr]
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            current_seq.append(arr[i])
        else:
            if len(current_seq) > len(best_seq):
                best_seq = current_seq
            current_seq = [arr[i]]
    if len(current_seq) > len(best_seq):
        best_seq = current_seq
    print(*best_seq)
```

```python
def smallest_with_one_swap(n_str):
    n_list = list(n_str)
    min_val = n_str
    for i in range(len(n_list)):
        for j in range(i + 1, len(n_list)):
            # Đổi chỗ
            n_list[i], n_list[j] = n_list[j], n_list[i]
            # Tránh số 0 ở đầu nếu số có nhiều hơn 1 chữ số
            if n_list != '0':
                current = "".join(n_list)
                if current < min_val:
                    min_val = current
            # Đổi lại để thử cặp tiếp theo
            n_list[i], n_list[j] = n_list[j], n_list[i]
    print(min_val)
```
