```python
def is_prime(k):
    if k < 2: return False
    for i in range(2, int(k**0.5) + 1):
        if k % i == 0: return False
    return True

a, b = map(int, input().split())
tong = 0
for i in range(a, b + 1):
    if is_prime(i):
        tong += i
print(tong)
```

```python
n = int(input())
d = 2
max_prime = 2
while n > 1:
    if n % d == 0:
        max_prime = d
        n //= d
    else:
        d += 1
print(max_prime)
```

```python
def reverse_num(n):
    return int(str(n)[::-1])

def is_palindrome(n):
    return str(n) == str(n)[::-1]

n = int(input())
buoc = 0
while not is_palindrome(n):
    n = n + reverse_num(n)
    buoc += 1
print(buoc, n)
```

```python
n = int(input())
res = []
i = 1
while i * i <= n:
    s = str(i * i)
    if len(set(s)) == len(s): 
        res.append(s)
    i += 1

if res:
    print(" ".join(res))
else:
    print("(no number)")
```

```python
def collatz_len(x):
    length = 1
    while x != 1:
        if x % 2 == 0: x //= 2
        else: x = 3 * x + 1
        length += 1
    return length

n = int(input())
max_L = 0
best_x = 0
for i in range(1, n + 1):
    L = collatz_len(i)
    if L > max_L:
        max_L = L
        best_x = i
print(best_x, max_L)
```

```python
tong_con = int(input())
tong_chan = int(input())

found = False
for g in range(tong_con + 1):
    c = tong_con - g
    if 2 * g + 4 * c == tong_chan:
        print(g, c)
        found = True
        break

if not found:
    print("invalid")
```

```python
def sum_divisors(k):
    s = 0
    for i in range(1, k):
        if k % i == 0: s += i
    return s

a = int(input())
b = int(input())
if sum_divisors(a) == b and sum_divisors(b) == a:
    print("true")
else:
    print("false")
```
