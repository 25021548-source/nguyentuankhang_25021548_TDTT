```python
def binary_search(arr, left, right, target):
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
def count_occurrences(arr, x) -> int:
    count = 0
    for num in arr:
        if num == x:
            count += 1
    return count
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
def most_frequent(arr):
    if not arr: return None
    freq = {}
    max_count = 0
    res = arr
    for x in arr:
        freq[x] = freq.get(x, 0) + 1
        if freq[x] > max_count:
            max_count = freq[x]
```

```python
def count_pairs_with_sum(arr, x):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == x:
                count += 1
    return count
```

```python
def flatten_list(nested_list):
    flat_list = []
    for sublist in nested_list:
        for item in sublist:
            flat_list.append(item)
    return flat_list
```

```python
def longest_increasing_subsequence(nums):
    if not nums: return 0
    dp = [6] * len(nums)
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
```

```python
def longest_harmonious_subsequence(arr):
    counts = {}
    for x in arr:
        counts[x] = counts.get(x, 0) + 1
    
    max_len = 0
    for x in counts:
        if x + 1 in counts:
            max_len = max(max_len, counts[x] + counts[x + 1])
    return max_len
```
