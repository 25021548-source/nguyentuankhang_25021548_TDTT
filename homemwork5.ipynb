```python
n, m, k = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(n)]
# Tổng hàng k
row_sum = sum(matrix[k])
# Tổng cột k
col_sum = sum(matrix[i][k] for i in range(n))
print(f"Tổng hàng {k}: {row_sum}, Tổng cột {k}: {col_sum}")
```

```python
def find_max_pos(lst):
    max_val = max(lst)
    return lst.index(max_val)
```

```python
def find_first_k(lst, k):
    try:
        return lst.index(k)
    except ValueError:
        return -1
```

```python
words = input().split()
d = {}
for i, word in enumerate(words):
    if word not in d:
        d[word] = []
    d[word].append(i)
result = {k: tuple(v) for k, v in d.items()}
print(result)
```

```python
def delete_less_than_x(lst, x):
    return [i for i in lst if i >= x]
```

```python
def move_zeros_to_front(lst):
    zeros = [i for i in lst if i == 0]
    others = [i for i in lst if i != 0]
    return zeros + others
```

```python
def sum_local_maxima(lst):
    total = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            total += lst[i]
    return total
```

```python
def find_pos_less_than_x(lst, x):
    positions = [i for i, val in enumerate(lst) if val < x]
    return positions if positions else -1
```

```python
def check_sequence(lst):
    tang = giam = True
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]: tang = False
        if lst[i] <= lst[i+1]: giam = False
    if tang: return "Tăng thực sự"
    if giam: return "Giảm thực sự"
    return "Vô danh"
```

```python
def find_missing_0_9(lst):
    missing = [i for i in range(10) if i not in lst]
    return missing
```
